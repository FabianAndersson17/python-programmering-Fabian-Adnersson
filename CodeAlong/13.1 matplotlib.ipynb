# Matplotlib
- MATLAB-style
- OOP approach

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5)
f = np.sin(x)

##print(x)
```

## MATLAB Style interface

```python
plt.figure(dpi = 100)
plt.plot(x, f)
plt.title("Sin(x)")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.figure(dpi = 100, figsize=(10, 4))
plt.subplot(1,2,1) ## (rows, cols, panel number)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.plot(x, np.tan(x))

plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
```

## OOP Approach

```python
fig = plt.figure(dpi=100)
## Ax objects can have a plot
ax = plt.axes()
print(fig)

ax.plot(x, f)
ax.set(xlabel = "x", ylabel = "y", title = "$\sin{x}$", xlim = (0,5));
```

```python
fig, ax = plt.subplots(1,2, dpi=100, figsize= (10,4))
rng = np.random.RandomState(42)
data_normal_dist = rng.randn(1000)

ax[0].hist(data_normal_dist, bins=30, alpha =.5)
ax[0].set(title = "Histogram of normal distibuted data", xlabel="x", ylabel="y")

bubble = ax[1].scatter(x=rng.randn(100), y=rng.randn(100), c = rng.randn(100), alpha = .7, cmap="cividis", s = 1000*rng.rand(100))
ax[1].set(title = "Bubble chart", xlabel="x", ylabel="y")
fig.colorbar(bubble)

plt.savefig("beutiful_plots.png", facecolor="White")
```
