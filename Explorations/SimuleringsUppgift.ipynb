# Rolig simulerings uppgift

```python
import math
import matplotlib.pyplot as plt
import random

simulations = 0
withinCircle = 0
def Euclidean_Distance(x, y):
    return math.sqrt((x-0)**2 + (y-0)**2)

radius = 1
print(Euclidean_Distance(0.5, 0.5))
print(Euclidean_Distance(1,1))
print(Euclidean_Distance(0.5, -0.5))

plt.axes().set_aspect(1)
for _ in range(1000):
    x = random.uniform(-radius, radius)
    y = random.uniform(-radius, radius)
    if Euclidean_Distance(x, y) <= radius:
        plt.scatter(x, y, c = "Green")
        withinCircle += 1
    else:
        plt.scatter(x, y, c = "Red")
    simulations += 1
plt.show()

amountInside = withinCircle/simulations
procentInside = amountInside*100
print(amountInside*4)
print(amountInside)
print(procentInside)
```
