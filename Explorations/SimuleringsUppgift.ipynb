# Rolig simulerings uppgift

```python
import math
import matplotlib.pyplot as plt
import random

simulations = 0
withinCircle = 0
def Euclidean_Distance(x, y):
    return math.sqrt((x-0)**2 + (y-0)**2)

print(Euclidean_Distance(0.5, 0.5))
print(Euclidean_Distance(1,1))
print(Euclidean_Distance(0.5, -0.5))

plt.axes().set_aspect(1)
for _ in range(1000):
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    if Euclidean_Distance(x, y) <= 1:
        plt.scatter(x, y, c = "Green")
        withinCircle += 1
    else:
        plt.scatter(x, y, c = "Red")
    simulations += 1
plt.show()

amountInside = withinCircle/simulations
print(amountInside)
```

```python
import matplotlib.pyplot as plt
import random
import math

withinCircle = 0
simulations = 0

plt.axes().set_aspect(1)
for _ in range(1000):
    x = random.uniform(-2, 2)
    y = random.uniform(-2, 2)
    if Euclidean_Distance(x, y) <= 2:
        plt.scatter(x, y, c = "Green")
        withinCircle += 1
    else:
        plt.scatter(x, y, c = "Red")
    simulations += 1
plt.show()

amountInside = withinCircle/simulations

print(amountInside)
procentInside = amountInside*100
print(f"{procentInside}%")
```
